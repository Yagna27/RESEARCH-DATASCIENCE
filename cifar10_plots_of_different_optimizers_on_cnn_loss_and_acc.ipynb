```python
fig,accuracy_all=plt.subplots()  # TRAIN-ACCURACY CURVE
epochs=[5,10,20,50,70,100]

acc_adam=[0.6156,0.7104,0.7709,0.8385,0.8755,0.9166]
acc_rmsprop=[0.6252,0.6048,0.6575, 0.6701,0.6455,0.6364]
acc_adadelta=[0.1461,0.1677,0.2175,0.2956,0.3626,0.4072]
acc_adagrad=[0.2728,0.3697,0.4231,0.4693,0.5039, 0.5315]
acc_yogi=[0.4988,0.5712,0.6354,0.6897,0.7150, 0.7409]
acc_nadam=[0.6302,0.7217,0.7812,0.8468,0.8827, 0.8854]

accuracy_all.plot(epochs,acc_adam,label="adam")
accuracy_all.plot(epochs,acc_rmsprop,label="rmsprop")
accuracy_all.plot(epochs,acc_adadelta,label="adadelta")
accuracy_all.plot(epochs,acc_adagrad,label="adagrad")
accuracy_all.plot(epochs,acc_yogi,label="yogi")
accuracy_all.plot(epochs,acc_nadam,label="nadam")

plt.xlabel("EPOCHS")
plt.ylabel("train ACCURACY")

accuracy_all.legend()
```

since we can see that , adam and nadam optimizers are almost equal but adam gives more train accuracy at the 100 epochs while nadam gives more at 60 and below.
